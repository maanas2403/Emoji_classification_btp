```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Read the text file line by line and store it in a list
with open('training_corpus_pos_final.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Create a pandas DataFrame with a single column containing the lines
df = pd.DataFrame({'Text': lines})
```

```python
# Read the text file line by line and store it in a list
with open('training_corpus_neg_final.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Create a pandas DataFrame with a single column containing the lines
df_2 = pd.DataFrame({'Text': lines})
```

```python
df_2
```

```python
with open('test_corpus_final.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Create a pandas DataFrame with a single column containing the lines
df_3 = pd.DataFrame({'Text': lines})
```

```python
df_3
```

```python
final_df=pd.concat([df,df_2,df_3])
final_df.reset_index(drop=True, inplace=True)
```

```python
final_df
```

```python
from deep_translator import GoogleTranslator
```

```python
def translate_to_english(text):# Assuming 'text_column' is the name of your column containing text in multiple languages
    try:
        translation = GoogleTranslator(source='bn', target='en').translate(text)
        print(f"Original text: {text}, Translated Text: {translation}")
        return translation
    except Exception as e:
        print(f"Error translating text for Text {text}: {e}")
        return None
```

```python
final_df['translated_text'] = final_df['Text'].apply(translate_to_english)
```

```python
final_df.to_csv('translated_text.csv', index=False, encoding='utf-8-sig')
```
