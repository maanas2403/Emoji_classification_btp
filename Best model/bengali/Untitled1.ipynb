```python
def translate_to_english(text):# Assuming 'text_column' is the name of your column containing text in multiple languages
    try:
        translation = GoogleTranslator(source='bn', target='en').translate(text)
        print(f"Original text: {text}, Translated Text: {translation}")
        return translation
    except Exception as e:
        print(f"Error translating text for Text {text}: {e}")
        return None
```
