```python
def replace_emoticons_with_emojis(text):
    emojis = []
    for emoticon, emoji in emoticon_to_emoji.items():
        if emoticon in text:
            text = text.replace(emoticon, emoji)
            emojis.append(emoji)
    return text,emojis 
```

```python
df['translated_text_emoji'], df['emojis'] = zip(*df['translated_text'].apply(replace_emoticons_with_emojis))
```

```python
df
```

```python
def remove_emojis(text):
    for emoji in emoticon_to_emoji.values():
        text = text.replace(emoji, '')
    return text
```

```python
df['translated_text_no_emoji'] = df['translated_text_emoji'].apply(remove_emojis)
```

```python
df
```

```python
emoji_to_number = {}
label_counter = 0
for emoji in emoticon_to_emoji.values():
    if emoji not in emoji_to_number:
        emoji_to_number[emoji] = label_counter
        label_counter += 1
```

```python
emoji_df = pd.DataFrame({'Emoji': list(emoji_to_number.keys()), 'Label': list(emoji_to_number.values())})
```

```python
emoji_df
```

```python
def replace_emojis_with_labels(emoji_list):
    return [emoji_df[emoji_df['Emoji'] == emoji]['Label'].values[0] for emoji in emoji_list]
```

```python
df['emoji_labels'] = df['emojis'].apply(replace_emojis_with_labels)
```

```python
new_df=df[['translated_text_no_emoji','emoji_labels']]
```

```python
from collections import Counter

# Assuming your DataFrame is named df and the column containing lists of emojis is 'emoji_column'
# Example DataFrame:
# df = pd.DataFrame({'emoji_column': [['üòä', 'üòÇ', 'üòä'], ['üòÇ', 'üòÇ', 'üòä', 'üòÜ'], ['üòÇ', 'üòÇ', 'üòä', 'üòä']]})

# Step 1: Flatten the lists of emojis into a single list
flat_list = [emoji for sublist in df['emojis'] for emoji in sublist]

# Step 2: Count the occurrences of each emoji
emoji_counts = Counter(flat_list)

# Step 3: Sort the emojis based on their frequencies
sorted_emojis = sorted(emoji_counts.items(), key=lambda x: x[1], reverse=True)

# Display the frequency list of emojis
for emoji, count in sorted_emojis:
    print(f"{emoji}: {count}")
```
